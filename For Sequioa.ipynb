## Classes

```
# classes are fancy objects 
# classes have attributes and methods

# attributes are unique qualities to that class
# methods are unique functions for that class

# in this example i am creating a pet
# pets have names (attribute) 
# pets can speak (method)

class Pet():
    """
    A class for cute lil pets

    Atrributes: 
        name: str, the name of your pet
        spayed: Boolean, default=False, True if your pet is spayed
                False if it is not
    
    Methods:
        get_name(): returns name of pet
        speak(text): prints {text} as if your pet said it
        is_spayed(): returns True statement if pet is spayed 
                     and False if it is not
        get_spayed(): spays your pet i.e. changes self.spayed to True
        has_babies(): your pet will have a baby and return a pet instance

    """

    def __init__(self, name):
        self.name = name
        self.spayed = False

    def get_name(self):
        """
        Returns name of pet

        Args():
            None
        Returns:
            self.name: str, name of pet
        """
        return self.name

    def speak(self, text):
        """
        prints pet_name says text

        Args(1):
            text: str, text you want your pet to say

        Returns:
            none
        """
        print(self.name, 'says', text)    
    
    def is_spayed(self):
        """
        Returns whether or not your pet is spayed.
        True for spayed, false for not spayed.

        Args():
            None
        
        Returns:
            self.spayed: Boolean, spayed status of your pet
        """
        return self.spayed

    def get_spayed(self):
        """
        Spays your pet if it is not already spayed.
        Will change self.spayed from False to True.

        Args():
            none
        Returns:
            none
        """
        self.spayed = True

    def has_babies(self):
        """
        Your pet will have a baby if it is not spayed. 
        Creates a new instance of a pet with the name baby. 

        Args():
            None
        Returns:
            baby: class Pet() instance 
        """

        if not self.spayed:
            return Pet('baby')
        else:
            raise ValueError('Pet is spayed. cannot have babies') 
            return None


```
