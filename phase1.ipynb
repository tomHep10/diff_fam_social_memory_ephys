```python
import os
import glob
import pandas as pd
#import boris_extraction as boris
```

```python
# Specify the directory where you want to search for .xlsx files
directory_path = r'D:\Phase 1\boris_data'

# Use glob to find all .xlsx files in the specified directory and its subdirectories
xlsx_files = glob.glob(os.path.join(directory_path, '*.xlsx'), recursive=True)

# Iterate through each Excel file and read its content
for file_path in xlsx_files:
    try:
        df = pd.read_excel(file_path)

        # Now you can work with the DataFrame 'df' representing the data from the Excel file
        # For example, you can print the first few rows of the DataFrame
        print(f"Content of {file_path}:")
        print(df.head())
        print("\n" + "="*50 + "\n")

    except pd.errors.EmptyDataError:
        # Handle empty Excel files if needed
        print(f"Empty file: {file_path}")
    except Exception as e:
        # Handle other exceptions
        print(f"Error reading {file_path}: {e}")
```
